```python
import pandas as pd
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("kkkkkkkk.xlsx")
```

```python
df['Profit_Loss'] = df['Bilans'] - df['Price SK (EUR/MWh)']
df.drop('Unnamed: 11', axis=1, inplace=True)
df.drop('Unnamed: 10', axis=1, inplace=True)
```

```python
df
```

```python

from io import StringIO


# Convert "Delivery day" to datetime and extract the month
df['Delivery day'] = pd.to_datetime(df['Delivery day'])
df['Month'] = df['Delivery day'].dt.month

# Filter the DataFrame for January (month=1)
january_df = df[df['Month'] == 1]

# Display the DataFrame for January
january_df
```

```python


# Calculate the minimum, maximum, and average of 'Bilans' for each 'Period number'
bilans_stats_january = january_df.groupby('Period number')['Bilans'].agg(['min', 'max', 'mean'])

bilans_stats_january
####### godziny STYCZEN
```
