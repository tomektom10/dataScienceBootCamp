```python
import pandas as pd
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Total_results_statystyka_024-01-01_2024-01-23.xlsx")
```

```python
df.columns
```

```python
# Dodanie kolumny 'Profit/Loss', która oblicza zysk lub stratę na podstawie 'DA price' i 'Bilans'
df['Profit_Loss'] = df['Bilans'] - df['DA price']
df  # Wyświetlenie pierwszych kilku wierszy zaktualizowanego DataFrame

```

```python

# Grupowanie danych wg 'Period' i obliczanie średniej, min, max dla 'DA price'
grouped = df.groupby('Period')['DA price'].agg(['mean', 'min', 'max'])

print(grouped)

# WARTOSCI DLA GODZIN Z ROKU 2024
```

```python


# Grupowanie danych wg 'Period' i obliczanie średniej, min, max dla 'DA price' oraz sumy dla 'Bilans'
grouped1 = df.groupby('Period').agg({'DA price': ['mean', 'min', 'max'], 'Bilans': 'mean'})

print(grouped1)

# DOŁOŻYŁEM JESZCZE  BILANS SREDNIA ZE WSZYSTKICH DNI
```

```python
# Obliczenie łącznego wyniku dla 'Profit/Loss' dla poszczególnych okresów z kolumny 'Period'
total_profit_loss_by_period = df.groupby('Period')['Profit_Loss'].sum()

total_profit_loss_by_period.head  # Wyświetlenie pierwszych kilku wyników
# ZYSKI/STRATY ŁĄCZNIE DLA POSZCZEGULNYCH GODZIN
```

```python
# Liczenie, ile razy 'Price Average' było większe od 'DA price' dla każdego okresu
count_higher_average_price = df[df['Price Average '] > df['DA price']].groupby('Period').size()

count_higher_average_price.head(24)  # Wyświetlenie wyników dla pierwszych kilku okresów

# TYLE RAZY DLA POSZCZEGULNYCH GODZIN CENA SREDNIA BYLA WYZSZA OD DA
```

```python
# Liczenie, ile razy 'Maximal price' było wyższe od 'DA price' o 5 dla każdego okresu
count_higher_maximal_price = df[df['Maximal price '] > df['DA price'] + 5].groupby('Period').size()

count_higher_maximal_price.head(24)  # Wyświetlenie wyników dla pierwszych kilku okresów

# TYLE RAZY MAXYMALAN CENA BYŁA WYZSZA OD DA O 5
```

```python
######## OD TEGO MOMENTU LICZE CZY USREDNIANIE MA SENS.  PONIZEJ WARTOŚCI O ILE SREDNIO CENA MAX BYŁA WIEKSZA OD DA DLA POSZCZEGULNYCH OKRESÓW

# Calculating the average difference between Maximal Price and DA Price for each Period
average_difference_by_period = df_profit_positive.groupby('Period').apply(lambda x: (x['Maximal price '] - x['DA price']).mean())
average_difference_by_period
```

```python
# Mapowanie wartości dodatkowych do okresów
additional_values_map = {
    1: 18.892857,
    2: 8.097778,
    3: 15.135556,
    4: 19.961111,
    5: 11.692222,
    6: 8.648571,
    7: 14.660000,
    8: 36.168889,
    9: 33.710000,
    10: 37.388333,
    11: 28.065714,
    12: 30.720714,
    13: 28.529167,
    14: 31.213333,
    15: 33.664615,
    16: 31.005000,
    17: 30.586000,
    18: 38.888571,
    19: 35.920000,
    20: 34.142727,
    21: 24.007500,
    22: 31.547273,
    23: 19.149167,
    24: 21.123333
}

# Dodawanie nowej kolumny do DataFrame
df_profit_positive['DA_plus_srednia_cena_max'] = df_profit_positive.apply(lambda x: x['DA price'] + additional_values_map.get(x['Period'], 0), axis=1)
df_profit_positive
######## DODAŁEM NOWĄ KOLUMNE GDZIE DO CENY DA DODAŁEM WARTOSCI WYZEJ
```

```python
df_profit_positive['Profit_Loss_po_dodaniu'] = df_profit_positive['Bilans'] - df_profit_positive['DA_plus_srednia_cena_max']
```

```python
###### ZOBACZYŁEM JAKI BYL ZYSK/STRATA JAKBYSMY KUPOWALI NIE PO CENIE DA ALE ZNACZNIE WYZEJ BO PO DA PLUS SREDNIA WARTOŚĆ MAX
```
