```python
import pandas as pd
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("kkkkkkkk.xlsx")
```

```python
df['Profit_Loss'] = df['Bilans'] - df['Price SK (EUR/MWh)']
```

```python
df.drop('Unnamed: 11', axis=1, inplace=True)
```

```python
df.drop('Unnamed: 10', axis=1, inplace=True)
```

```python
df
```

```python
########## 5 ###########
```

```python


# Wybieranie wierszy, gdzie 'Maximal price' jest większe od 'Price SK' o 5
max_wieksze_5 = df[df['Maximal price (EUR/MWh)'] > (df['Price SK (EUR/MWh)'] + 5)]

max_wieksze_5
```

```python

# Obliczenie ogólnego zysku z kolumny "Profit_Loss"
total_profit = max_wieksze_5['Profit_Loss'].sum()

total_profit

##### zysk tylko z tego gdzie MAX był wiekszy od DA o 5
```

```python


# Dodanie nowej kolumny, gdzie do 'Price SK (EUR/MWh)' dodajemy 5
max_wieksze_5['Price SK Plus 5'] = max_wieksze_5['Price SK (EUR/MWh)'] + 5

max_wieksze_5.head()

```

```python
max_wieksze_5.columns
```

```python
max_wieksze_5['Profit_Loss_plus5'] = max_wieksze_5['Bilans'] - max_wieksze_5['Price SK Plus 5']
```

```python
max_wieksze_5
```

```python
total_profit_plus5 = max_wieksze_5['Profit_Loss_plus5'].sum()
```

```python
total_profit_plus5
```
