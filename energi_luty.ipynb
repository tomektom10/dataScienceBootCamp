```python
import pandas as pd
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("kkkkkkkk.xlsx")
```

```python
df['Profit_Loss'] = df['Bilans'] - df['Price SK (EUR/MWh)']
df.drop('Unnamed: 11', axis=1, inplace=True)
df.drop('Unnamed: 10', axis=1, inplace=True)
```

```python
df
```

```python



# Convert 'Delivery day' to datetime if it's not already
df['Delivery day'] = pd.to_datetime(df['Delivery day'])

# Filter for February
february_data = df[df['Delivery day'].dt.month == 2]
```

```python
february_data
```

```python

from io import StringIO


# Filter data where 'Minimal price (EUR/MWh)' is at least 5 less than 'Price SK (EUR/MWh)'
luty_min_mniejsze = february_data[february_data['Minimal price (EUR/MWh)'] <= february_data['Price SK (EUR/MWh)'] - 5]

luty_min_mniejsze
### CENA MINIALNA 0 5 MNIEJSZA NIZ PRICE SK.
```

```python


# Grouping by 'Period number' and summarizing 'Profit_Loss'
ZYSKI_GODZINY = february_data.groupby('Period number')['Profit_Loss'].sum().reset_index()

ZYSKI_GODZINY
```

```python
# Summing the total Profit/Loss
ZYSKI_GODZINY_total = february_data['Profit_Loss'].sum()
ZYSKI_GODZINY_total
```

```python


# Sorting the data by 'Profit_Loss' and getting the top 30 rows
top_30_profit_loss_luty = february_data.sort_values(by='Profit_Loss', ascending=False).head(30)

top_30_profit_loss_luty.head(5)
```

```python
# Filtering the data where 'Price SK (EUR/MWh)' is not less than 25 and not greater than 170
february_data_odcinam_najgorsze = february_data[(february_data['Price SK (EUR/MWh)'] >= 25) & (february_data['Price SK (EUR/MWh)'] <= 160)]

february_data_odcinam_najgorsze
```

```python
ZYSKI_GODZINY_odciete = february_data_odcinam_najgorsze['Profit_Loss'].sum()
ZYSKI_GODZINY_odciete
##333 ZYSKI JAK ODCIAŁEM CENY MNIEJSZE OD 25 I WIEKSZE OD 160
```

```python

# Splitting the DataFrame into profit and loss DataFrames
df_profit_luty = ZYSKI_GODZINY[ZYSKI_GODZINY['Profit_Loss'] > 0]
df_loss_luty = ZYSKI_GODZINY[ZYSKI_GODZINY['Profit_Loss'] <= 0]

df_profit_luty
####### godziny z zyskiem od longa ############
```
