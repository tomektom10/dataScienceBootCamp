```python
wybrane_przedzialy = pd.concat([szort_data_above_100, szort_data_50_to_100, long_data_50_to_100, long_data_under50 ], ignore_index=True)
```

```python
wybrane_przedzialy
```

```python
suma_wartosci_wybrane_przedzialy =wybrane_przedzialy["Profit/loss"].sum()
```

```python
#CAŁY ZYSK DLA GRANIA WYBRANYCH PRZEDZIALOW
```

```python
suma_wartosci_wybrane_przedzialy
```

```python

# Inicjalizujemy zmienne
max_loss_series_length = 0
current_loss_series_length = 0
start_index = 0

# Iterujemy po DataFrame
for index, row in wybrane_przedzialy.iterrows():
    if row['Profit/loss'] < 0:
        current_loss_series_length += 1
    else:
        if current_loss_series_length > max_loss_series_length:
            max_loss_series_length = current_loss_series_length
            start_index = index - current_loss_series_length
        current_loss_series_length = 0

# Sprawdzamy dla ostatniej serii strat, jeśli jest najdłuższa
if current_loss_series_length > max_loss_series_length:
    start_index = len(wybrane_przedzialy) - current_loss_series_length

# Wyodrębniamy odpowiednie wiersze
max_loss_series_wybrane_przedzialy = wybrane_przedzialy.iloc[start_index:start_index + max_loss_series_length]

# Wyświetlamy wyniki
print("Najdłuższa seria strat:")
print(max_loss_series_wybrane_przedzialy)

```

```python
suma_wartosci_loss_in_row =max_loss_series_wybrane_przedzialy["Profit/loss"].sum()
```
