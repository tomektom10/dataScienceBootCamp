```python
wybrane_przedzialy = pd.concat([szort_data_above_100, szort_data_50_to_100, long_data_50_to_100, long_data_under50 ], ignore_index=True)
```

```python
wybrane_przedzialy
```

```python
suma_wartosci_wybrane_przedzialy =wybrane_przedzialy["Profit/loss"].sum()
```

```python
#CAŁY ZYSK DLA GRANIA WYBRANYCH PRZEDZIALOW
```

```python
suma_wartosci_wybrane_przedzialy
```

```python

# Inicjalizujemy zmienne
max_loss_series_length = 0
current_loss_series_length = 0
start_index = 0

# Iterujemy po DataFrame
for index, row in wybrane_przedzialy.iterrows():
    if row['Profit/loss'] < 0:
        current_loss_series_length += 1
    else:
        if current_loss_series_length > max_loss_series_length:
            max_loss_series_length = current_loss_series_length
            start_index = index - current_loss_series_length
        current_loss_series_length = 0

# Sprawdzamy dla ostatniej serii strat, jeśli jest najdłuższa
if current_loss_series_length > max_loss_series_length:
    start_index = len(wybrane_przedzialy) - current_loss_series_length

# Wyodrębniamy odpowiednie wiersze
max_loss_series_wybrane_przedzialy = wybrane_przedzialy.iloc[start_index:start_index + max_loss_series_length]

# Wyświetlamy wyniki
print("Najdłuższa seria strat:")
print(max_loss_series_wybrane_przedzialy)

```

```python
suma_wartosci_loss_in_row =max_loss_series_wybrane_przedzialy["Profit/loss"].sum()
```

```python
suma_wartosci_loss_in_row
```

```python
wybrane_przedzialy['Cum. Return'] = wybrane_przedzialy['Profit/loss'].cumsum()
```

```python
wybrane_przedzialy
```

```python
import matplotlib.pyplot as plt
```

```python
wybrane_przedzialy['Cum. Return'] = wybrane_przedzialy['Profit/loss'].cumsum()

# Znajdujemy maksymalną cenę ujemną
max_negative_price = wybrane_przedzialy['Cum. Return'].min()

# Rysujemy wykres dla pierwszych 200 wyników
plt.figure(figsize=(15, 8))
plt.plot(df['Cum. Return'].head(50), label='Cumulative Return', marker='o', linestyle='-')

# Zaznaczamy maksymalną cenę ujemną
plt.axhline(y=max_negative_price, color='red', linestyle='--', label='Max Negative Price')

# Ustawiamy zakres osi y od -500
plt.ylim(-200, max(wybrane_przedzialy['Cum. Return'].head(200)) + 500)

# Dodajemy etykiety i tytuł
plt.xlabel('Numer wiersza')
plt.ylabel('Cumulative Return')
plt.title('Kumulacyjny Zwrot (Pierwsze 200 wyników)')

# Dodajemy legendę
plt.legend()

# Dodajemy etykiety liczbowe ujemne na osi y z większymi krokami
yticks_range = range(-500, int(max(wybrane_przedzialy['Cum. Return'].head(200))) + 500, 100)
plt.yticks(yticks_range)

# Wyświetlamy wykres
plt.show()
```
