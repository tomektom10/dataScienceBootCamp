```python
import pandas as pd
import pandas as pd
import numpy as np
df = pd.read_excel("ROBERT_model_moje_sprawdzenie.xlsx")
```

```python
############################LICZE TAM GDZIE DOSTALISMY LONGI####################################################
```

```python


# Tworzenie nowego DataFrame z warunkiem logicznym
df_longowy = df[df['po longa 0,5'] > df['DA']]

df_longowy
```

```python
#################### licze to co zamkniete po 20 ##############
```

```python

df_longowy_zamkniete_20 = df_longowy[df_longowy['Maximal price (EUR/MWh)'] > df_longowy['DA'] + 20]




df_longowy_zamkniete_20
```

```python
######ZYSKI Z TEGO ZAMYKANIA PO 20 POZIOMACH 2234 * 20 = 44680
```

```python
df_longowy_NIE_zamkniete_20 = df_longowy[df_longowy['Maximal price (EUR/MWh)'] <= df_longowy['DA'] + 20]
df_longowy_NIE_zamkniete_20
```

```python
# Obliczanie zysku/straty (Profit/Loss)
df_longowy_NIE_zamkniete_20['Profit/Loss'] = df_longowy_NIE_zamkniete_20['bilans'] - df_longowy_NIE_zamkniete_20['DA']

df_longowy_NIE_zamkniete_20
```

```python
suma_longowa_tam_gdzie_niedali = df_longowy_NIE_zamkniete_20['Profit/Loss'].sum()
suma_longowa_tam_gdzie_niedali
```

```python
######################### LICZE TAM GDZIE DOSTALISMY SZORTY #####################
```

```python
# Tworzenie nowego DataFrame z warunkiem logicznym
df_szortowy = df[df['po szorta 0,5'] < df['DA']]

df_szortowy
```

```python
################## licze to co zamknieto po 20 ###########################
```

```python

df_szortowy_zamkniete_20 = df_szortowy[df_szortowy['Minimal price (EUR/MWh)'] < df_szortowy['DA'] - 20]




df_szortowy_zamkniete_20
```

```python
################### ZYSK Z MAKNIETYCH SZORTOWYCH PO 20 LEVELACH  2534 * 20 = 50680
```

```python
df_szortowy_NIE_zamkniete_20 = df_szortowy[df_szortowy['Minimal price (EUR/MWh)'] >= df_szortowy['DA'] - 20]




df_szortowy_NIE_zamkniete_20
```

```python
df_szortowy_NIE_zamkniete_20['Profit/Loss'] = df_szortowy_NIE_zamkniete_20['DA'] - df_szortowy_NIE_zamkniete_20['bilans']
df_szortowy_NIE_zamkniete_20
```

```python
suma_szortowa_tam_gdzie_niedali = df_szortowy_NIE_zamkniete_20['Profit/Loss'].sum()
suma_szortowa_tam_gdzie_niedali
```
