```python
import pandas as pd
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("kkkkkkkk.xlsx")
```

```python
df['Profit_Loss'] = df['Bilans'] - df['Price SK (EUR/MWh)']

df
```

```python
from io import StringIO



# Konwersja kolumny 'Delivery day' do formatu daty
df['Delivery day'] = pd.to_datetime(df['Delivery day'])

# Wyodrębnienie danych za styczeń 2023
df_january = df[df['Delivery day'].dt.month == 1]

df_january
```

```python
zyski_strata_styczen = df_january['Profit_Loss'].sum()
zyski_strata_styczen
# ZYSKI STRATA STYCZEN
```

```python


# Filtering rows where 'Price SK' is between 0 and 50
od0_do50_df = df_january[(df_january['Price SK (EUR/MWh)'] >= 0) & (df['Price SK (EUR/MWh)'] <= 50)]

# Calculating the total Profit/Loss for these rows
total_profit_loss = od0_do50_df['Profit_Loss'].sum()
total_profit_loss
# PRZEDZIAL 0-50 ZYSKI /STRATA Z GRANIA LONGOW
```

```python
###### 10 NAJWIKESZYCH STRAT W STYCZNIU 2023
```

```python
# Creating a new DataFrame with the 10 largest losses based on the 'Profit_Loss' column
df_largest_losses = df_january.nsmallest(10, 'Profit_Loss')

df_largest_losses.reset_index(drop=True, inplace=True)
df_largest_losses
```

```python
######### wpisałem 10 największych strat z dodatnim bilansem myślalem ze beda jaies straty od szorta ale tutaj też same od longa
```

```python
# Filtering the DataFrame for rows where 'Bilans' is positive and then finding the 10 largest losses
df_positive_balance_losses = df_january[df_january['Bilans'] > 0].nsmallest(10, 'Profit_Loss')

df_positive_balance_losses.reset_index(drop=True, inplace=True)
df_positive_balance_losses
```

```python
###### w tym DF BILANS WIEKSZY NIZ PRICE WIEC STRATA Z SZORTA
```

```python
# Creating a new DataFrame where 'Bilans' is greater than 'Price SK'
df_bilans_greater_than_price_sk = df_january[df_january['Bilans'] > df_january['Price SK (EUR/MWh)']]

df_bilans_greater_than_price_sk.reset_index(drop=True, inplace=True)
df_bilans_greater_than_price_sk
```

```python
# Re-calculating the profit or loss with the correct interpretation
# When selling at 'Price SK' and buying back at 'Bilans', the profit or loss is calculated as:
# Profit or Loss = Price SK - Bilans

df_bilans_greater_than_price_sk['Profit_loss_od_szorta'] = df_bilans_greater_than_price_sk['Price SK (EUR/MWh)'] - df_bilans_greater_than_price_sk['Bilans']

df_bilans_greater_than_price_sk
```

```python

# Finding the 10 largest losses (smallest Profit_Loss values)
largest_losses = df_bilans_greater_than_price_sk.nlargest(20, 'Profit_Loss')

largest_losses.reset_index(drop=True, inplace=True)
largest_losses
######## NAJWIKESZKE STRATY W STYCZNIU 2023 OD SZORTA.....
```

```python


# Filtering for records where 'Price SK' does not exceed 200
DA_ponizej_200 = df_bilans_greater_than_price_sk[df_bilans_greater_than_price_sk['Price SK (EUR/MWh)'] <= 150]

# Finding the 20 largest losses (most negative values) in 'Profit_loss_od_szorta'
largest_losses = DA_ponizej_200.nsmallest(20, 'Profit_loss_od_szorta')

largest_losses.reset_index(drop=True, inplace=True)
largest_losses
######### najwikesze straty kiedy FIX był poniżej 150. praktycznie duze straty najczesciej sa przy wysokich cenach
```
