```python
import pandas as pd
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("kkkkkkkk.xlsx")
```

```python
df['Profit_Loss'] = df['Bilans'] - df['Price SK (EUR/MWh)']

df
```

```python
from io import StringIO



# Konwersja kolumny 'Delivery day' do formatu daty
df['Delivery day'] = pd.to_datetime(df['Delivery day'])

# Wyodrębnienie danych za styczeń 2023
df_january = df[df['Delivery day'].dt.month == 1]

df_january
```

```python
zyski_strata_styczen = df_january['Profit_Loss'].sum()
zyski_strata_styczen
# ZYSKI STRATA STYCZEN
```

```python


# Filtering rows where 'Price SK' is between 0 and 50
od0_do50_df = df_january[(df_january['Price SK (EUR/MWh)'] >= 0) & (df['Price SK (EUR/MWh)'] <= 50)]

# Calculating the total Profit/Loss for these rows
total_profit_loss = od0_do50_df['Profit_Loss'].sum()
total_profit_loss
# PRZEDZIAL 0-50 ZYSKI /STRATA Z GRANIA LONGOW
```
