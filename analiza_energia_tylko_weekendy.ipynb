```python
import pandas as pd
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("kkkkkkkk.xlsx")
```

```python
df = df.drop(df.index[-1])
```

```python
suma = df['ZYSK/STRATA'].sum()
```

```python


# Konwersja kolumny "Delivery day" na typ daty
df['Delivery day'] = pd.to_datetime(df['Delivery day'])

# Dodanie kolumny z dniem tygodnia
df['Day of Week'] = df['Delivery day'].dt.dayofweek

# Wybór weekendów (sobota: 5, niedziela: 6)
weekend_df = df[df['Day of Week'].isin([5, 6])]

# weekend_df zawiera teraz tylko dane z sobót i niedziel
```

```python
df
```

```python
weekend_df
```

```python



# Wybieranie tylko potrzebnych kolumn
selected_columns = ['Delivery day', 'Period number', 'Period', 'Traded Price Average (EUR/MWh)',
                    'Maximal price (EUR/MWh)', 'Bilans', 'Price SK (EUR/MWh)', 'ZYSK/STRATA', 'Day of Week']
new_df = weekend_df[selected_columns]

new_df.head(5)  # Pokazuje kilka pierwszych wierszy nowego DataFrame

```

```python
len(new_df)
```

```python
suma_weekendy = new_df['ZYSK/STRATA'].sum()
```

```python
suma_weekendy
```

```python
# Grupowanie danych po kolumnie 'Period number' i obliczanie sumy 'ZYSK/STRATA' dla każdej grupy
# ROZBICIE NA GODZINY CALY ROK TYLKO WEEKENDY !!!
grouped_data_cały_rok = df.groupby('Period number')['ZYSK/STRATA'].sum()

grouped_data_cały_rok

```

```python
# Dodanie kolumny z miesiącem
# ROZBICIE NA GODZINY MIESIACE TYLKO WEEKENDY
df['Month'] = df['Delivery day'].dt.month

# Grupowanie danych po miesiącu i 'Period number', a następnie obliczanie sumy 'ZYSK/STRATA' dla każdej grupy
grouped_monthly_data_godziny_miesiace = df.groupby(['Month', 'Period number'])['ZYSK/STRATA'].sum()

grouped_monthly_data_godziny_miesiace

```

```python
# Wyświetlenie całej zawartości DataFrame grouped_monthly_data_godziny_miesiace
# Zakładając, że grouped_monthly_data_godziny_miesiace to nazwa DataFrame, którą chcesz wyświetlić
# grouped_monthly_data_godziny_miesiace jest już zdefiniowany z poprzednich kroków

grouped_monthly_data_godziny_miesiace.reset_index()

```
