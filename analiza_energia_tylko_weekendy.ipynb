```python
df = df.drop('Unnamed: 12', axis=1)
```

```python


# Konwersja kolumny "Delivery day" na typ daty
df['Delivery day'] = pd.to_datetime(df['Delivery day'])

# Dodanie kolumny z dniem tygodnia
df['Day of Week'] = df['Delivery day'].dt.dayofweek

# Wybór weekendów (sobota: 5, niedziela: 6)
weekend_df = df[df['Day of Week'].isin([5, 6])]

# weekend_df zawiera teraz tylko dane z sobót i niedziel
```

```python
# SUMA ZYSKÓW WEEKEN !!!!!!
suma_weekend = weekend_df['$$$$'].sum()
suma_weekend
```

```python
# Grupowanie danych po kolumnie 'Period number' i obliczanie sumy '$$$$' dla każdej grupy
# ROZBICIE NA GODZINY CALY ROK TYLKO WEEKENDY !!!
grouped_data_cały_rok = weekend_df.groupby('Period number')['$$$$'].sum()

grouped_data_cały_rok

```
