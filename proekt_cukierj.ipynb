```python
def oblicz_ceny_flaja(df1, df2, kod1, kod2, cena1, cena2, day_month):

  wynik_dict = {"Date":[], "cena_flaja1":[] }
  df_wynik = pd.DataFrame()
  for year in [2023, 2022, 2021, 2020]:
    specyfic_date = str(year) + "-" + day_month
    specyfic_date = datetime.datetime.strptime(specyfic_date, '%Y-%m-%d')
    print(specyfic_date)
    df_row1 = df1[df1[kod1] == specyfic_date]
    print(df_row1)

    df_row2 = df2[df2[kod2] == specyfic_date]
    print(df_row2)
    cena_flaja = []
    if len(df_row1) > 0 and len(df_row2) >0:

      value = df_row1[cena1].values[0]
      cena_flaja = value - df_row2[cena2].values[0]


      wynik_dict["Date"].append(specyfic_date)
      wynik_dict["cena_flaja1"].append(cena_flaja)


      df_wynik = pd.DataFrame(wynik_dict)
  return df_wynik
```

```python
import calendar
month = "02"
```

```python
def create_days_for_month(month, year = 2023):
  numdays = calendar.monthrange(year, int(month))[-1]
  print(calendar.monthrange(year, int(month)))
  days_list = [f'{month}-{day:02d}' for day in range(1, numdays + 1)]
  return days_list
```

```python
def oblicz_ceny_flaja_dla_miesiaca(month, df1, df2, kod1 , kod2 , ceny1, ceny2):
  days_list = create_days_for_month(month)
  srednia_df = []
  for month_day in days_list:
    df= oblicz_ceny_flaja(df1, df2, kod1 , kod2 , ceny1, ceny2, month_day )
    if len(df) > 0:
      srednia_df.append({"month_day": month_day, "cena_flaja_srednia": df["cena_flaja1"].mean()})
  return srednia_df
```

```python
oblicz_ceny_flaja_dla_miesiaca("01", df1, df2, "SBES1H23" , "SBES1K23" , "Unnamed: 1", "Unnamed: 4" )
```

```python
srednie_ceny_rok = []
for month_i in range(1,13):
  month = f"{month_i:02d}"
  cena_flaja_miesiac = oblicz_ceny_flaja_dla_miesiaca(month, df1, df2, "SBES1H23" , "SBES1K23" , "Unnamed: 1", "Unnamed: 4" )
  srednie_ceny_rok += cena_flaja_miesiac

```
