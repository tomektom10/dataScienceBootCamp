```python
import pandas as pd
df = pd.read_excel("Cukry_wszystkie.xlsx")
df
df.columns
import numpy as np
import datetime
```

```python
df
```

```python
df1 = df[["SBES1H23", "Unnamed: 1"]].iloc[1:]
df1
```

```python
df2 = df[["SBES1K23", "Unnamed: 4"]].iloc[1:]
df2
import datetime

df3 = df[["SBES1N23", "Unnamed: 7"]].iloc[1:]
df3
```

```python
# przerobiłem tak zeby w przypadku braku danych wyswietlało puste.
```

```python
days_of_month = ["04-01", "04-02", "04-03", "04-04",  "04-05", "02-24", "02-23", "02-25", "02-26", "02-27", "11-01", "11-02", "11-03", "11-04", "11-05", "11-06", "11-07", "11-08", "11-09", "11-10", "11-11", "11-12", "11-13", "11-14"]  # Lista dwóch różnych dni

wynik_dict = {"Date": [], "cena_flaja1": [], "cena_flaja2": []}

for day_month in days_of_month:
    temp_dict = {"Date": [], "cena_flaja1": [], "cena_flaja2": []}

    for year in [2023, 2022, 2021, 2020, 2019, 2018, 2017, 2016, 2015, 2014]:
        specyfic_date = str(year) + "-" + day_month
        specyfic_date = datetime.datetime.strptime(specyfic_date, '%Y-%m-%d')
        print(specyfic_date)

        df_row1 = df1[df1["SBES1H23"] == specyfic_date]
        print(df_row1)

        df_row2 = df2[df2["SBES1K23"] == specyfic_date]
        print(df_row2)

        cena_flaja = []
        for i, value in enumerate(df_row1['Unnamed: 1'].values):
            roznica = value - df_row2["Unnamed: 4"].values[i]
            cena_flaja.append(roznica)

        if len(cena_flaja) == 1:
            temp_dict["Date"].append(specyfic_date)
            temp_dict["cena_flaja1"].append(cena_flaja[0])
            temp_dict["cena_flaja2"].append(np.nan)
        elif len(cena_flaja) == 2:
            temp_dict["Date"].append(specyfic_date)
            temp_dict["cena_flaja1"].append(cena_flaja[0])
            temp_dict["cena_flaja2"].append(cena_flaja[1])
        else:
            temp_dict["Date"].append(specyfic_date)
            temp_dict["cena_flaja1"].append(np.nan)
            temp_dict["cena_flaja2"].append(np.nan)

    wynik_dict["Date"].extend(temp_dict["Date"])
    wynik_dict["cena_flaja1"].extend(temp_dict["cena_flaja1"])
    wynik_dict["cena_flaja2"].extend(temp_dict["cena_flaja2"])

df_wynik = pd.DataFrame(wynik_dict)
```

```python
df_wynik
```

```python

df_wynik['Date'] = pd.to_datetime(df_wynik['Date'])  # Konwersja kolumny 'Date' na format daty

df_wynik['Year'] = df_wynik['Date'].dt.year
df_wynik['Rest_of_Date'] = df_wynik['Date'].dt.strftime('%m-%d')  # Pozostawienie tylko części miesiąca i dnia
```

```python
df_wynik
```

```python
df_wynik['Date'] = pd.to_datetime(df_wynik['Date'])
april_data = df_wynik[df_wynik['Date'].dt.month == 4]

# Oblicz średnią, medianę, wartość maksymalną i minimalną kwecień
mean_april_price = april_data['cena_flaja1'].mean()
median_april_price = april_data['cena_flaja1'].median()
max_april_price = april_data['cena_flaja1'].max()
min_april_price = april_data['cena_flaja1'].min()

df_wynik['Date'] = pd.to_datetime(df_wynik['Date'])
luty_data = df_wynik[df_wynik['Date'].dt.month == 2]

# Oblicz średnią, medianę, wartość maksymalną i minimalną luty
mean_luty_price = luty_data['cena_flaja1'].mean()
median_luty_price = luty_data['cena_flaja1'].median()
max_luty_price = luty_data['cena_flaja1'].max()
min_luty_price = luty_data['cena_flaja1'].min()


df_wynik['Date'] = pd.to_datetime(df_wynik['Date'])
listopad_data = df_wynik[df_wynik['Date'].dt.month == 11]

# Oblicz średnią, medianę, wartość maksymalną i minimalną listopad
mean_listopad_price = listopad_data['cena_flaja1'].mean()
median_listopad_price = listopad_data['cena_flaja1'].median()
max_listopad_price = listopad_data['cena_flaja1'].max()
min_listopad_price = listopad_data['cena_flaja1'].min()

```

```python
data1 = {
    'Miesiąc': ['Kwiecień', 'Luty', "Listopad"],
    'Średnia': [mean_april_price, mean_luty_price, mean_listopad_price],
    'Mediana': [median_april_price, median_luty_price, median_listopad_price],
    'Wartość maksymalna': [max_april_price, max_luty_price, max_listopad_price],
    'Wartość minimalna': [min_april_price, min_luty_price, min_listopad_price]
}

df_wynik1 = pd.DataFrame(data1)
df_wynik1.set_index('Miesiąc', inplace=True)  # Ustawienie kolumny "Miesiąc" jako indeksu

print(df_wynik1)
```
