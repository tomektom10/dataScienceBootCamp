```python
import pandas as pd
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Kwiecein_do_liczenia.xlsx")
```

```python
df.columns
```

```python
df['Profit/Loss'] = df['Bilans'] - df['Price SK (EUR/MWh)']
```

```python
df
```

```python
total_profit_loss = df['Profit/Loss'].sum()
total_profit_loss
```

```python
# Konwersja kolumny 'Delivery day' na typ daty
df['Delivery day'] = pd.to_datetime(df['Delivery day'])

# Dodanie kolumny identyfikującej dzień tygodnia (0=Poniedziałek, 6=Niedziela)
df['Day of Week'] = df['Delivery day'].dt.dayofweek

# Filtracja dni tygodnia (0-4 to Poniedziałek-Piątek)
weekday_df = df[df['Day of Week'] < 5]

# Filtracja weekendów (5-6 to Sobota-Niedziela)
weekend_df = df[df['Day of Week'] >= 5]

# Usunięcie pomocniczej kolumny 'Day of Week' jeśli nie jest już potrzebna
weekday_df = weekday_df.drop(columns=['Day of Week'])
weekend_df = weekend_df.drop(columns=['Day of Week'])

```

```python
weekday_df
```

```python


# Obliczenie różnicy między 'Maximal price' a 'Price SK'
weekday_df['Price Difference'] = weekday_df['Maximal price (EUR/MWh)'] - weekday_df['Price SK (EUR/MWh)']

# Filtracja rekordów, gdzie różnica była większa lub równa 20
filtered_df = weekday_df[weekday_df['Price Difference'] >= 20]

# Grupowanie danych po kolumnie 'Period number' i zliczanie rekordów w każdej grupie
dni_tygodnia_pogrupowane = filtered_df.groupby('Period number').size().reset_index(name='Count')
```

```python
dni_tygodnia_pogrupowane
```

```python
period_1_count = (weekday_df['Period number'] == 1).sum()
```

```python
period_1_count
```

```python
# Obliczenie różnicy między 'Price SK' a 'Minimal price'
weekday_df['Price Difference'] = weekday_df['Price SK (EUR/MWh)'] - weekday_df['Minimal price (EUR/MWh)']

# Filtracja rekordów, gdzie różnica była większa lub równa 20
filtered_df_szort = weekday_df[weekday_df['Price Difference'] >= 20]

# Grupowanie danych po kolumnie 'Period' i zliczanie rekordów w każdej grupie
pogrupowane_szort = filtered_df_szort.groupby('Period').size().reset_index(name='Count')
```

```python
pogrupowane_szort
```

```python

# Obliczenie różnicy między 'Maximal price' a 'Price SK'
weekend_df['Price Difference'] = weekend_df['Maximal price (EUR/MWh)'] - weekend_df['Price SK (EUR/MWh)']

# Filtracja rekordów, gdzie różnica była większa lub równa 20
filtered_df_weekend = weekend_df[weekend_df['Price Difference'] >= 20]

# Grupowanie danych po kolumnie 'Period number' i zliczanie rekordów w każdej grupie
weekend_pogrupowane = filtered_df_weekend.groupby('Period number').size().reset_index(name='Count')
```
